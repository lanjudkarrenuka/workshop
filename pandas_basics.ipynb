```python
 import pandas as pd
 import numpy as np
```

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series (country)
```

```python
runs = [13,24,56,78,100]
runs_ser = pd.Series (runs)
runs_ser
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['marks','english','maths','science']
marks_ser = pd.Series (marks,index=subjects)
marks_ser
```

```python
marks_2 = {'marks':67,'english':57,'maths':89,'science':100}
marks_2_ser = pd.Series (marks_2)
marks_2_ser
```

```python
marks_1 = {'marks':67,
           'english':57,
           'maths':89,
           'science':100}
marks_1_ser = pd.Series (marks_1)
marks_1_ser
```

```python
# setting a name
marks = [67,57,89,100]
subjects = ['marks','english','maths','science']
marks_ser = pd.Series (marks,index=subjects,name='marks')
marks_ser
```
